```python
import pandas as pd

# Read all CSVs
june_df = pd.read_csv("hardcovers_nyt_bestsellers_june_2025.csv")
july_df = pd.read_csv("hardcovers_nyt_bestsellers_july_2025.csv")
aug_df = pd.read_csv("hardcovers_nyt_bestsellers_august_2025.csv")
sept_df = pd.read_csv("hardcovers_nyt_bestsellers_sept_2025.csv")
oct_df = pd.read_csv("hardcovers_nyt_bestsellers_oct_2025.csv")
nov_df = pd.read_csv("hardcovers_nyt_bestsellers_nov_2025.csv")
```

```python
# Combine them into a single DataFrame
all_months_df = pd.concat([june_df, july_df, aug_df, sept_df, oct_df, nov_df], ignore_index=True)
```

```python
# Sort by date and rank
all_months_df['date'] = pd.to_datetime(all_months_df['date'])
all_months_df = all_months_df.sort_values(by=['date', 'rank'])
```

```python
# Save combined CSV
all_months_df.to_csv("hardcovers_nyt_bestsellers_2025.csv", index=False)
print("Combined CSV saved successfully!")
```
